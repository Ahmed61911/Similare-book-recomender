### Creation du Pivot table (matrice)

```python
book_pivot = final_rating.pivot_table(columns = 'user_id', index = 'title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
data_long = book_pivot.stack().reset_index()
data_long.columns = ['title', 'user_id', 'rating']

# Filter out the zero ratings if necessary
data_long = data_long[data_long['rating'] > 0]

# Create a mapping of book titles to their indices
title_to_index = {title: index for index, title in enumerate(book_pivot.index)}

# Replace book titles with their indices
data_long['title_index'] = data_long['title'].map(title_to_index)

# Create the scatter plot using indices instead of titles
plt.figure(figsize=(20, 12))
plt.scatter(data_long['user_id'], data_long['title_index'], c=data_long['rating'], s=20, cmap='viridis', alpha=0.6)
plt.colorbar(label='Rating')
plt.xlabel('User ID')
plt.ylabel('Book Title Index')
plt.title('User Ratings for Books')
plt.show()
```

### Systeme de scoring pour trier les resultats

```python
def calculate_score(row):
    # Example scoring function based on average rating and number of ratings
    # You can adjust the weights as needed
    score = (row['rating'] * 0.7) + (row['num_of_rating'] * 0.3)
    return score

def recommend_book1(book_name, num_recommendations=5):
    if book_name not in book_pivot.index:
        print("Book not found in the dataset.")
        return
    
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=num_recommendations + 1)
    
    # Create a DataFrame to store the recommended books along with their distances
    recommended_books = []
    for i in range(len(suggestions[0])):
        book_index = suggestions[0][i]
        book_title = book_pivot.index[book_index]
        dist = distance[0][i]
        avg_rating = final_rating[final_rating['title'] == book_title]['rating'].mean()
        num_of_rating = final_rating[final_rating['title'] == book_title]['num_of_rating'].values[0]
        recommended_books.append((book_title, dist, avg_rating, num_of_rating))
    
    # Convert to DataFrame
    recommended_df = pd.DataFrame(recommended_books, columns=['title', 'distance', 'rating', 'num_of_rating'])
    
    # Calculate scores
    recommended_df['score'] = recommended_df.apply(calculate_score, axis=1)
    
    # Sort by score
    recommended_df = recommended_df.sort_values(by='score', ascending=False)
    
    # Print recommendations
    print(f"Top {num_recommendations} recommendations for '{book_name}':")
    for index, row in recommended_df.iloc[1:num_recommendations + 1].iterrows():
        print(f"Title: {row['title']}, Score: {row['score']:.2f}, Distance: {row['distance']:.2f}, Avg Rating: {row['rating']:.2f}, Number of Ratings: {row['num_of_rating']}")
        
```
